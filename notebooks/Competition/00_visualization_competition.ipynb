## 00. Visualization

```python
import pandas as pd
import dtale

# Function to show df in D-Tale
def show_df_in_dtale(df):
    instance = dtale.show(df)
    instance.kill()
    d = dtale.show(df, enable_custom_filters=True)
    print(f"D-Tale running on: {d._main_url}")
```

#### **Processed for training**

```python
df = pd.read_csv("../../data/Competition/processed_for_training/competition_processed_for_training_week_h6_t_plus_1.csv", sep=";")
df
```

```python
df.columns
```

```python
# Columns that end with "_scaled"
scaled_cols = [col for col in df.columns if col.endswith("_scaled")]

# Same column names without "_scaled"
original_cols_to_drop = [col.removesuffix("_scaled") for col in scaled_cols]

# Keep only those that actually exist in the original df
original_cols_to_drop = [col for col in original_cols_to_drop if col in df.columns]

# Drop them from df
df = df.drop(columns=original_cols_to_drop)
```

```python
show_df_in_dtale(df)
```
