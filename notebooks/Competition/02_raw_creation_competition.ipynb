## 02. Raw creation

Saves the original data in raw

```python
import pandas as pd
import dtale

# Function to show df in D-Tale
def show_df_in_dtale(df):
    instance = dtale.show(df)
    instance.kill()
    d = dtale.show(df, enable_custom_filters=True)
    print(f"D-Tale running on: {d._main_url}")
```

```python
#df_products = pd.read_excel("../../data/competition/raw/competition - Aquiles_products.xlsx")
df = pd.read_csv("../../data/competition/raw/Week 8 - 2024-06-03 - Sales.csv")
df
```

```python
df_melted = df.melt(
    id_vars=['Store', 'Product'],
    var_name='datetime',
    value_name='value'
)

df_melted['datetime'] = pd.to_datetime(df_melted['datetime'])

df_melted
```

```python
df_melted.to_parquet(
    "../../data/competition/raw/competition_raw.parquet",
    index=False,
    engine="pyarrow",
)
```

```python
show_df_in_dtale(df_melted)
```
