## 02. Raw creation

Saves the original data in raw

```python
import pandas as pd
import dtale

# Function to show df in D-Tale
def show_df_in_dtale(df):
    instance = dtale.show(df)
    instance.kill()
    d = dtale.show(df, enable_custom_filters=True)
    print(f"D-Tale running on: {d._main_url}")
```

```python
#df_products = pd.read_excel("../../data/Lecta/raw/LECTA - Aquiles_products.xlsx")
df_transactions = pd.read_excel("../../data/Lecta/raw/LECTA - Aquiles_transactions.xlsx")
df_transactions
```

```python
df_transactions.value_counts('warehouse')
```

```python
df_transactions.to_parquet(
    "../../data/Lecta/raw/lecta_raw.parquet",
    index=False,
    engine="pyarrow",
)
```

```python
show_df_in_dtale(df_transactions)
```
